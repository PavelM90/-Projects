## Training of the initial object detection model and image recognition model

```python
# Import libraries
from ultralytics import YOLO
import torch
```

```python
# Generate an instance of a YOLO v10 model
model = YOLO('yolov10l.pt')
```

```python
# Move model to GPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)
```

```python
# Train the model on football match video cut
results = model.predict('clip.mp4', save = True)
```
